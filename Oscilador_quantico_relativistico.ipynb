```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
gm=0
hw=1
def parte_Z1_dependenteN(n, w, h, m, T, c):
  termo1 = gm*(n**2) #-2*func_mi(w, m, c)*(n**2)
  termo2 = (hw - 2*gm)*n #(h*w - (2*func_mi(w, m, c)))*n
  z1_2 = -beta(T)*(termo1 + termo2)
  return np.exp(z1_2)

def func_mi(w, m, c):
  mi = 3*((h*w)/2)**2/(8*fun_eV(m, c))
  return 0

def fun_eV(m, c):
  e_J = m*c**2
  e = e_J * 6.241509*10**(18)
  return e

def beta(T):
  K = 1 # 8.61733e-5
  return 1/(K*T)

def parte_constante_Z1(T, h, w, m, c):
  return  np.exp((-beta(T)*(hw/2 - gm) ) )#np.exp((-beta(T))*(((h*w)/2) - func_mi(w, m, c)))
```

```python
#CONSTANTES (definir aqui)
c = 1 #1299792458 #velocidade da luz em m/s
w = 1
h = 1
#h = 6.58211928*10**(-16) #constante de planck reduzida em eV
```

```python
#temperaturas = np.linspace(0.1,100, 100)
temperaturas = np.linspace(1,10, 100)
lista_n = np.linspace(0,500, 501)

Z1_2 = []
#m = [1, 10, 10**2, 10**3, 10**4]
m = 1
for T in temperaturas:
  soma_n = []
  soma = 0
  for n in lista_n:
    x = parte_Z1_dependenteN(n, w, h, m, T, c)
    soma_n.append(x)
    #print(x)
  soma = sum(soma_n)
  #print(soma)
  Z1_2.append(parte_constante_Z1(T, h, w, m, c)*soma)
```

```python
plt.title(f'Variação de Z1_2 em relação à temperatura')
plt.plot(temperaturas, Z1_2)
plt.xlabel('Temperatura (K)')
plt.ylabel('Z1_2')
plt.show()
```

```python
plt.title(f'Variação da função de partição em relação à temperatura')
plt.plot(temperaturas, Z1, label='Relativístico', color='purple')
plt.plot(temperaturas, Z1_2, label = 'Sólido de Einstein',color='deeppink')
plt.xlabel('Temperatura (K)')
plt.ylabel('Função de partição')
plt.legend()
plt.show()
```

### Cálculo da Energia de Helmholtz

```python
def helmholtz(T, z):

  print(z)
  print(np.log(z))

  F = - (beta(T)**(-1)) * np.log(z)

  return F
```

```python
energias_helmholtz = []
for t, z in zip(temperaturas, Z1):
  #print(t)
  #print(z)
  energias_helmholtz.append(helmholtz(t, z))

plt.title("", fontsize = 14)
plt.plot(temperaturas, energias_helmholtz)
plt.xlabel("Temperatura", fontsize = 14)
plt.ylabel("Energia de Helmholtz", fontsize = 14)
plt.show()
```

```python
energias_helmholtz_2 = []
for t, z in zip(temperaturas, Z1_2):
  print(t)
  print(z)
  energias_helmholtz_2.append(helmholtz(t, z))

plt.title("", fontsize = 14)
plt.plot(temperaturas, energias_helmholtz_2)
plt.xlabel("Temperatura", fontsize = 14)
plt.ylabel("Energia de Helmholtz", fontsize = 14)
plt.show()
```

```python
plt.title("", fontsize = 14)
plt.plot(temperaturas, energias_helmholtz, color='purple', label= 'Relativístico')
plt.plot(temperaturas, energias_helmholtz_2, color='deeppink', label = 'Sólido de Einstein')
plt.xlabel("Temperatura", fontsize = 14)
plt.ylabel("Energia de Helmholtz", fontsize = 14)
plt.title("Energia de Helmholtz em função da temperatura")
plt.legend()
plt.show()
```

### Cálculo da Energia média

```python
#Calculo da derivada

def derivada_xy(x_vals, y_vals):
    avg_x_vals = []
    deriv_vals = []
    for i in range(len(x_vals) - 1):
        deriv = ( y_vals[i + 1] - y_vals[i] ) / ( x_vals[i + 1] - x_vals[i] ) *(-1)
        avg_x = ( x_vals[i + 1] + x_vals[i] ) / 2

        avg_x_vals.append(avg_x)
        deriv_vals.append(deriv)

        kb = 1 #1.380649e-23

        deriv_certo = []
        for i in deriv_vals:
            d = i*kb
            deriv_certo.append(d)

        x = avg_x_vals
        y = deriv_certo

    return x,y

# Valores da temperatura e cálculo do ln de Z

#T_vals = [3,5,8,3,1,2,6,7]
#Z_lista = [1,2,3,4,5,6,7,8]

print(Z1)

ln_Z = []
for i in Z1:
    i = abs(i)
    Z_ln  = np.log(i)
    ln_Z.append(Z_ln)

x_medio, derivada = derivada_xy(temperaturas,ln_Z)

plt.title(f'Energia média do sistema')
plt.plot(x_medio, derivada, 'purple')
plt.xlabel("Temperaturas")
plt.ylabel("Energia média")
plt.show()
```
